```python
from Lens import ConvexLens
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib notebook
```

# 设置透镜并画出透镜形状

```python
lens = ConvexLens(width=0.1, height=1, refractive=1.25) # 设置透镜厚度，高度，折射率
lens.xrange = (-2, 2)

plt.figure(figsize=(10, 5))
lens.plot_axis()
lens.plot_lens()
plt.show()
```

# 光线经过凸透镜会发生偏折

```python
x0, y0, k0, v0 = -15, 0.9, 0, 1 # 光线起点坐标，斜率，方向

plt.figure(figsize=(10, 5))
lens.get_path(x0, y0, k0, v0, x_end=15)
lens.xrange = (-15, 15)
lens.plot_path()
lens.reset()
plt.show()
```

# 验证平行光经过薄凸透镜可以汇聚于一点

```python
x0 = -15
y0s = np.linspace(-0.9, 0.9, 10)
k0, v0 = 0, 1

plt.figure(figsize=(10, 5))
for y0 in y0s:
    lens.get_path(x0, y0, k0, v0, x_end=15)
lens.xrange = (-15, 15)
lens.plot_path()
lens.reset()
plt.grid()
plt.show()
```

## 检验厚凸透镜（完全球形透镜）对于光线的汇聚情况

```python
lens_sp = ConvexLens(width=1, height=1, refractive=1.25)
x0 = -3
y0s = np.linspace(-0.9, 0.9, 10)
k0, v0 = 0, 1

plt.figure(figsize=(10, 10/3))
for y0 in y0s:
    lens_sp.get_path(x0, y0, k0, v0, x_end=3)
lens_sp.xrange = (-3, 3)
lens_sp.plot_path()
lens_sp.reset()
plt.show()
```

# 验证焦点出射的光线经过薄凸透镜会形成平行光/光路可逆

```python
x0, y0 = -10, 0
y1 = np.linspace(-0.9, 0.9, 10)
k0s = (y0 - y1) / x0
v0 = 1

plt.figure(figsize=(10, 5))
for k0 in k0s:
    lens.get_path(x0, y0, k0, v0, x_end=15)
lens.xrange = (-15, 15)
lens.plot_path()
lens.reset()
plt.grid()
plt.show()
```

# 验证焦点内出射的光线经过薄凸透镜不能会聚在一起

```python
x0, y0 = -5, 0
y1 = np.linspace(-0.9, 0.9, 10)
k0s = (y0 - y1) / x0
v0 = 1

plt.figure(figsize=(10, 5))
for k0 in k0s:
    lens.get_path(x0, y0, k0, v0, x_end=15)
lens.xrange = (-15, 15)
lens.plot_path()
lens.reset()
plt.grid()
plt.show()
```

# 验证二倍焦距处出射的光线经过薄凸透镜会在另一边的二倍焦距处会聚

```python
x0, y0 = -20, 0
y1 = np.linspace(-0.9, 0.9, 10)
k0s = (y0 - y1) / x0
v0 = 1

plt.figure(figsize=(10, 5))
for k0 in k0s:
    lens.get_path(x0, y0, k0, v0, x_end=25)
lens.xrange = (-25, 25)
lens.plot_path()
lens.reset()
plt.grid()
plt.show()
```

# 验证薄凸透镜二倍焦距处的成像规律

```python
x0 = -20
y0s = np.linspace(0, 0.5, 3)
y1 = np.linspace(-0.9, 0.9, 7)
v0 = 1

plt.figure(figsize=(10, 5))
for y0 in y0s:
    k0s = (y0 - y1) / x0
    for k0 in k0s:
        lens.get_path(x0, y0, k0, v0, x_end=25)
lens.xrange = (-25, 25)
lens.plot_path()
lens.reset()
plt.grid()
plt.show()
```

# 验证薄凸透镜二倍焦距以外的成像规律

```python
x0 = -25
y0s = np.linspace(0, 0.5, 3)
y1 = np.linspace(-0.9, 0.9, 7)
v0 = 1

plt.figure(figsize=(10, 5))
for y0 in y0s:
    k0s = (y0 - y1) / x0
    for k0 in k0s:
        lens.get_path(x0, y0, k0, v0, x_end=25)
lens.xrange = (-25, 25)
lens.plot_path()
lens.reset()
plt.grid()
plt.show()
```
